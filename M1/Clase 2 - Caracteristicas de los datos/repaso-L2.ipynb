```python
archivo = open('Emisiones_CO2.csv','r',encoding='latin-1')
```

```python
dicc_emisiones = {
    'cod_pais': [],
    'nom_pais': [],
    'region': [],
    'año': [],
    'co2(kt)': [],
    'co2_percapita': []    
}
```

```python
next(archivo)
for linea in archivo:
    linea = linea.strip()
    campo = linea.split('|')
    dicc_emisiones['cod_pais'].append(campo[0])
    dicc_emisiones['nom_pais'].append(campo[1])
    dicc_emisiones['region'].append(campo[2])
    dicc_emisiones['año'].append(campo[3])
    dicc_emisiones['co2(kt)'].append(campo[4])
    dicc_emisiones['co2_percapita'].append(campo[5])
    
```
