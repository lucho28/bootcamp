```python
import random as r
import numpy as np

def ListaDesordenada(desorden,cantidad):
    
    lista = list(range(0,cantidad))
    desorden = int(cantidad * desorden / 100)
    
    while (desorden > 0):
        i = r.randint(0,cantidad-1)
        j = r.randint(0,cantidad-1)
        aux = lista[i]
        lista[i] = lista[j]
        lista[j] = aux
        desorden-=1

    return lista
```

```python

ListaDesordenada(100,5000)
```

```python
lista_desordenada = ListaDesordenada(100,6000)
```

```python
def selection_sort(lista):
    
    for i in range(0,len(lista)):
        for j in range(i+1,len(lista)):
            if lista[j]>lista[i]:
                aux = lista[j]
                lista[j] = lista[i]
                lista[i] = aux
    
    return lista
        
```

```python
import time
start_time = time.time()
selection_sort(lista_desordenada)
total_time = time.time() - start_time
print(format(total_time))
```

```python
lista = ListaDesordenada(80,6000)
```

```python
def bubble_sort(lista):
    
    for i in range(1,len(lista)):
        for j in range(0,len(lista)):
            if lista[i] < lista[j]:
                aux = lista[j]
                lista[j] = lista[i]
                lista[i] = aux
    
    return lista
```

```python
bubble_sort(lista)
```

```python
start_time = time.time()
bubble_sort(lista)
final_time = time.time() - start_time
print("Llevo {} segundos en total.".format(total_time))
```
